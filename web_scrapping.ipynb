```python
# Instalasi yang diperlukan
!apt-get update
!apt install -y chromium-chromedriver
!pip install -U selenium webdriver-manager

# Import library yang diperlukan
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--remote-debugging-port=9222")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("start-maximized")
chrome_options.add_argument("disable-infobars")
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("--window-size=1920,1080")

# Setup WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # Buka halaman
    driver.get("https://www.abgi.or.id/")

    # Tunggu beberapa detik untuk memastikan konten dimuat
    time.sleep(5)  # Anda dapat menyesuaikan waktu tunggu ini

    # Ambil konten dari elemen dengan ID "map-biogas"
    map_biogas_element = driver.find_element(By.ID, "map-biogas")
    map_biogas_content = map_biogas_element.get_attribute('innerHTML')

    # Tampilkan konten
    print(map_biogas_content)

finally:
    # Tutup driver
    driver.quit()
```

```python
import json
import pandas as pd
import re
```

```python
# Step 1: Use regex to extract JSON from JavaScript variables
potensi_match = re.search(r'var dataPotensi = (\[.*?\]);', map_biogas_content)
instalasi_match = re.search(r'var dataInstalasi = (\[.*?\]);', map_biogas_content)

if potensi_match and instalasi_match:
    # Load JSON data
    dataPotensi = json.loads(potensi_match.group(1))
    dataInstalasi = json.loads(instalasi_match.group(1))

    # Step 2: Convert JSON data to DataFrames
    potensi_df = pd.DataFrame(dataPotensi)
    instalasi_df = pd.DataFrame(dataInstalasi)

    # Step 3: Save DataFrames as CSV files
    potensi_df.to_csv('dataPotensi.csv', index=False)
    instalasi_df.to_csv('dataInstalasi.csv', index=False)

    # Display the DataFrames to verify
    print("Data Potensi:")
    print(potensi_df.head())
    print("\nData Instalasi:")
    print(instalasi_df.head())
else:
    print("Failed to find dataPotensi or dataInstalasi in the HTML content.")
```
